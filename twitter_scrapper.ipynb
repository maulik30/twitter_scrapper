```python
#Importing modules
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
```

```python
#Setting begin and end dates for tweets
begin_date = dt.date(2020,3,1)
end_date = dt.date(2020,3,31)
```

```python
#setting other arguments for query_tweets
limit=1000
lang= 'english'
user= 'realDonaldTrump'
```

```python
#extracting tweets
tweets = query_tweets("corona", begindate=begin_date, enddate=end_date, limit=limit, lang=lang)
```

```python
#creating a dataframe from the tweets
df= pd.DataFrame(t.__dict__ for t in tweets)
df.head()
```
